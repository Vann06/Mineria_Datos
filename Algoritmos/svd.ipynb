```python
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler

try:
    df = pd.read_csv('../movies_2026.csv', encoding='latin-1')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: movies_2026.csv not found.")
```

```python
#Columnas numericas relevantes
numeric_features = ['budget', 'revenue', 'runtime', 'popularity', 'voteAvg', 'voteCount', 
                    'genresAmount', 'productionCoAmount', 'productionCountriesAmount', 
                    'actorsAmount', 'castWomenAmount', 'castMenAmount']

#Llenar datos nulos con 0
df_num = df[numeric_features].fillna(0)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_num)

svd = TruncatedSVD(n_components=5, random_state=42)
svd.fit(df_scaled)

print('Porcentaje de Varianza Explicada por componente:', svd.explained_variance_ratio_)
print("\nDesglose de pesos del Componente (Las variables que más lo influyen):")

for i, comp in enumerate(svd.components_):
    top_features = sorted(zip(numeric_features, comp), key=lambda x: abs(x[1]), reverse=True)[:3]
    top_str = " | ".join([f"{feat}: {peso:.3f}" for feat, peso in top_features])
    print(f'Componente {i+1}: {top_str}')
```

## Interpretacion de los componentes

Componente 1: Son las películas a las que se les mete muchísimo dinero, terminan recaudando una fortuna y todo el mundo habla de ellas.

Componente 2: Son películas que se graban en un montón de países diferentes y tienen muchísimos actores hombres, pero que curiosamente terminan ganando menos dinero en la taquilla.

Componente 3: Son películas con elencos masculinos gigantes y grabadas por todo el mundo, pero que, por extraño que parezca, son pagadas y controladas por muy poquitas empresas productoras.

Componente 4: Son películas con repartos bastante pequeñitos donde la gran mayoría de los personajes son mujeres, las cuales suelen durar un poquito menos de lo normal y se enfocan en muy pocos géneros a la vez.

Componente 5: Son estas mega producciones que contratan a un batallón gigante de actores y actrices simplemente para llenar la pantalla de gente, logrando así llamar un poco más la atención y popularidad del público.
